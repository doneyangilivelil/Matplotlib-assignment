```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show
%matplotlib inline
```

```python
titanic_train = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

'1. Create a pie chart presenting the male/female proportion

```python
sums = titanic_train.pclass.groupby(titanic_train.sex).sum()
axis('equal');
pie(sums, labels=sums.index);
show()
```

'2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.style.use('classic')
```

```python
titanic_train.head(2)
```

```python
titanic = titanic_train.dropna(subset=['sex'])
mapping = {'male' : 'blue', 'female' : 'red'}
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, c=titanic['sex'].map(mapping), label='FarePaid vs Age :M and F')
plt.show()
```
